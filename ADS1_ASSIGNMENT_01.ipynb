```python
"""Line graph for Hig, Low stock price """

#using a function to call variables for line plot
def line_graph():
    '''Line graph for netflix stock high and low price'''
    fig, ax = plt.subplots(figsize=[8,5])
    ax.plot(data.Date,data.High,color='g')
    ax.plot(data.Date,data.Low,color='r')
    ax.yaxis.set_major_formatter('${x:,.0f}')
   
    font1 = {'family':'serif','color':'black','size':12}
    plt.legend(['High','Low'])
    plt.xlabel('Year & Month')
    plt.ylabel('High($)')
    plt.title("Netflix stock price change from 2023 to 2024 ",fontdict = font1)
    plt.savefig('Netflix stock price change over the months.png')
line_graph()
```

```python
#using a function to call the code inside the function 
def hist():
    '''Histogram using seaborn for netflix stock High and Low'''
    
    plt.figure(figsize= [13,5])
    font1 = {'family':'serif','color':'black','size':12}
    plt.subplot(1,2,1)
    sns.histplot(data=data,x=data.High,kde=True,bins=12,line_kws={'color':'crimson', 'lw': 2, 'ls': ':'}, color='green')
    plt.title('Netflix Stock High',fontdict = font1)
    plt.xlabel('High($)')

    plt.subplot(1,2,2)
    sns.histplot(data=data,x=data.Low,kde=True,bins=12, line_kws={'color':'crimson', 'lw': 2, 'ls': ':'},color ='red')
    plt.title('Netflix Stock Low',fontdict = font1)
    plt.xlabel('Low($)')

    plt.savefig('Stock gain and reduction.png')
hist()
```

```python
#using a function to call the code inside the function 
def box_plot():
    
    '''Box plot using seaborn for the Netflix stock'''
    
   
    fig,ax = plt.subplots(figsize=[8,6])
    sns.boxplot(data[['High','Low','Open','Close']],showmeans=True,meanline=True,palette="Paired",meanprops={"color": "black"})
    ax.yaxis.set_major_formatter('${x:,.0f}')
    font1 = {'family':'serif','color':'black','size':12}
    plt.title('Comparison Between High, Low and Open, Close from 2023 to 2024',fontdict=font1)
    plt.ylabel('Stock Price')
    plt.savefig('Comparison Between High, Low and Open, Close.png')
box_plot()
```

```python
"""Function call"""
line_graph()
hist()
box_plot()
```
